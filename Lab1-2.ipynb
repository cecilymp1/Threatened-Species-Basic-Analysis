One figure with three histograms (arranged horizontally) of the average skill level of
students, one for each category:

- Use `plt.subplots()` to initialize a figure with three Axes objects arranged
horizontally. Make it 7" wide and 3" high.
- For each student in the current class, compute the average of points that the
student reported across all skills in a given category. Do this separately for each
category: coding (code_), statistics (stat_) and spatial data (spat_).
- For each of the three categories, create a histogram (not barplots) that shows
the distribution of these averages.
- Plot each histogram on one of the three axes objects. Make sure the x-axis
extent covers the right range (1-3). Add an informative title to each histogram.
Save a 150dpi version of the figure at this filepath:

````python
#initializing three axes objects horizontally
fig, axes = plt.subplots(1, 3)
fig.set_size_inches(7, 3)


#Category: coding (code_)
code_cols = [v for v in d.columns if v.startswith('code_')]
d[code_cols]

#category: stat_
stat_cols = [v for v in d.columns if v.startswith('stat_')]
d[stat_cols]

#category: spat_
spat_cols = [v for v in d.columns if v.startswith('spat_')]
d[spat_cols]

#computing the average and then creating a histogram
d[code_cols].mean(1).plot(kind='hist', ax=axes[0])
axes[0].set_title('Code Experience', fontsize=12, pad=15)   
d[stat_cols].mean(1).plot(kind='hist', ax=axes[1])
axes[1].set_title('Stat Experience', fontsize=12, pad=15)   
d[spat_cols].mean(1).plot(kind='hist', ax=axes[2])
axes[2].set_title('Spatial Experience', fontsize=12, pad=15)

filepath = 'Deliverables/L1/2_ee508_skills_hist.png'
plt.savefig(filepath, bbox_inches='tight', dpi=150)
````

### One figure with two barplots of students' software skills and interests of the current Fall 2023 class
- Use plt.subplots() to initialize a figure with two axes objects (arranged vertically)
that is 4" wide and 6" high.
- For each software type (sw_), compute the average skill level reported across all
students of the current class (i.e., a single value per software type). Create a
horizontal bar plot that shows the class' average skill level for each software
type and plot it on the upper axes. Add a title, make it look nice, and choose the
correct x-axis extent (1-5). Do the same for each self-reported interest (int_) and
plot it on the lower axes.
- The title of the lower plot might overlap with the labels of the y axis of the
upper plot. The quickest way to resolve this issue is a matplotlib shortcut which
automatically chooses a spacing which resolves such overlaps:
plt.tight_layout()
- Save a 150dpi version at:
Deliverables/L1/2_ee508_sw_int.png

````python
#initializing two axes objects arranged vertically
fig, axes = plt.subplots(2,1, figsize=(4,6))

#Category: software type: sw_
sw_cols = [v for v in d.columns if v.startswith('sw_')]
d[sw_cols]

average_sw = d[sw_cols].mean()

#computing the average and then creating a barplot


axes[0].barh(sw_cols, average_sw)
axes[0].set_xlabel('Average Skill Level')
axes[0].set_ylabel('Software type')
axes[0].set_title('Average Skill Level by Software Type')
axes[0].set_xlim(1, 5)




#Category: interest: int_
int_cols = [v for v in d.columns if v.startswith('int_')]
d[int_cols]

average_int = d[int_cols].mean()

#computing the average and then creating a barplot


axes[1].barh(int_cols, average_int)
axes[1].set_xlabel('Average Interest')
axes[1].set_ylabel('Choices for choosing to take this class')
axes[1].set_title('Average Interest')
axes[1].set_xlim(1, 5)
plt.tight_layout()
plt.show()

#saving the image
filename = 'Deliverables/L1/2_ee508_sw_int.png'
plt.savefig(filename,bbox_inches='tight', dpi=150)
````
